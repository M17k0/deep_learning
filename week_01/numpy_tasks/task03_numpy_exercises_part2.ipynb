````python
print('array of 10 zeros')
print(f'Python: {[0 for _ in range(10)]}')
print(f'NumPy: {np.zeros(10)}')

print('\narray of 10 ones')
print(f'Python: {[1 for _ in range(10)]}')
print(f'NumPy: {np.ones(10)}')

print('\narray of 10 fives')
print(f'Python: {[5 for _ in range(10)]}')
print(f'NumPy: {np.ones(10) * 5}')

print('\narray of 10 zeroes, the fifth value of which is 1')
print(f'Python: {[0]*4 + [1] + [0]*5}')
print(f'NumPy: {np.eye(1, 10, 4).ravel()}')

print('\narray of all the integers from 10 to 50')
print(f'Python: {list(range(10, 51))}')
print(f'NumPy: {np.arange(10, 51)}')

print('\narray of all the even integers from 10 to 50')
print(f'Python: {[x for x in range(10, 51) if x % 2 == 0]}')
print(f'NumPy: {np.arange(10, 51, 2)}')

arr = [1, 2, 3]
reversed_arr = arr[::-1]
np_arr = np.array(arr)
reversed_np_arr = np_arr[::-1]

pure_matrix = [[3 * i + j for j in range(3)] for i in range(3)]
print(f'\nPython 3x3 matrix:\n{pure_matrix}')
numpy_matrix = np.arange(9)
print(f'\nNumPy 3x3 matrix:\n{numpy_matrix}')
````

#### Task 10: Reproducibility

When training models, it's often important to generate the same random numbers every time the application is run, even if the machine changes. This ability helps us remove the possibility of creating the situation: *Well, it works on my machine...*.

Our code should work on all machines and produce the same results across different runs.

Go through the [documentation](https://numpy.org/doc/stable/reference/module_structure.html) and answer the below question.

How can we ensure that we always get the same random numbers (even if we switch machines)? Give two examples.

> **Your answer**: We can ensure that we always get the same random numbers by setting a random seed, so every time the the same sequence of numers is produced.


Example 1: Using `np.random.seed()`

````python
np.random.seed(42)
num = np.random.rand()
print(num)
````

Example 2: Using a `Generator` instance

````python
rng = np.random.default_rng(42)
num = rng.random()
print(num)
````
